```python
pivotA = nbrA.pivot_table(index='id_client', columns='id_carte_fidélité', values='Nombre_A_ou_MA')

pivotA = pivotA.fillna(0)

pivotA.head()
```

```python
scaler = StandardScaler()
pivotA_scaled = scaler.fit_transform(pivotA)
```

```python
# Initialize the KNN model with cosine similarity
model_knn = NearestNeighbors(metric='cosine', n_neighbors=5)  # Considering 6 neighbors, including the client itself

model_knn.fit(pivotA_scaled)
```

```python
# Calculate the similarity indices
distances, indices = model_knn.kneighbors(pivotA_scaled)
```

```python
clt_sim = pd.DataFrame(indices, columns=['client0', 'client1', 'client2', 'client3', 'client4'])

clt_sim.head()
```

```python
# Create a dictionary to map client IDs to names from nbrA DataFrame
id_to_name_mapping = nbrA.set_index('id_client')['nom_prenom'].to_dict()

# Define the number of top similar clients to extract
N = 5  # Change this value as needed

# Dictionary to store top similar clients for each client
top_similar_clients = {}

# Loop through each column (client) in the similarity table
for column in client_names_df.columns:
    # Get the current client ID
    client_id = int(column.replace('client', ''))

    # Get the current client name
    client_name = id_to_name_mapping.get(client_id, "Unknown")

    # Get the top N similar clients by sorting the column
    top_n_similar_clients = client_names_df[column].sort_values(ascending=False).head(N + 1)  # Include the client itself

    # Exclude the current client from the list
    top_n_similar_clients = top_n_similar_clients[top_n_similar_clients.index != column]

    # Map client IDs to names
    similar_client_names = [id_to_name_mapping.get(int(client), "Unknown") for client in top_n_similar_clients.index]

    # Store the top similar clients in the dictionary
    top_similar_clients[client_name] = similar_client_names

# Print top similar clients for each client
for client, similar_clients in top_similar_clients.items():
    print(f"Client: {client}")
    print(f"Top Similar Clients: {similar_clients}")
    print()
```

```python
import re
import unicodedata



def remove_non_alphanumeric(input_string):
    if isinstance(input_string, str):
        cleaned_string = re.sub(r'[^a-zA-Z0-9\s]', '', unicodedata.normalize('NFD', input_string).encode('ascii', 'ignore').decode('utf-8'))
        return cleaned_string
    else:
        return input_string

```

```python
nb=nbrA.copy()
# Apply the function to the 'nom_prenom' column in your dataset

nb['nom_prenom_cleaned'] = nb['nom_prenom'].apply(remove_non_alphanumeric)

# Display the cleaned dataset
nb.head()
```
