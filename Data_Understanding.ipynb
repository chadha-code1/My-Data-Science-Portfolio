top n function by secteur

`````python
import pandas as pd

def get_top_clients_by_sector(nbrA, selected_sector, top_n):
    # Filter the data for the selected sector
    selected_sector_data = nbrA[nbrA['Secteur'] == selected_sector]

    # Group the data by 'id_client' and 'Secteur', and calculate the mean spending for each client in the sector
    grouped_data = selected_sector_data.groupby(['id_client', 'Secteur'])['Nombre_A_ou_MA'].mean()

    # Sort the grouped data by mean spending in descending order
    sorted_data = grouped_data.sort_values(ascending=False)

    # Get the top N clients who spend the most in the selected sector
    top_n_clients = sorted_data.head(top_n)

    # Reset index and convert to DataFrame
    top_clients_df = top_n_clients.reset_index()

    return top_clients_df

# Get user input for sector and number of top clients
selected_sector = input("Enter the sector you want: ")
top_n = int(input("Enter the number of top clients: "))

# Get the top clients for the selected sector
top_clients = get_top_clients_by_sector(nbrA, selected_sector, top_n)

# Print the top clients' information
print(top_clients)
`````

**Rating the clients**

`````python
depensesMax = pd.DataFrame(nbrA.groupby('id_client')['Nombre_A_ou_MA'].mean())

depensesMax.head()
`````

`````python
# Add depensesMax of Nombre_A_ou_MA to dataset

depensesMax['nombre_depenses'] = pd.DataFrame(df.groupby('id_client')['Nombre_A_ou_MA'].count())

depensesMax.head()
`````

Top 10 best clients  by CBA


`````python
c = depensesMax['Nombre_A_ou_MA'].mean()
c
`````

`````python
m = depensesMax['nombre_depenses'].quantile(0.85)
m
`````

`````python
def weighted_rating(x, m=m, c=c):
    a = x['nombre_depenses']
    b = x['Nombre_A_ou_MA']
    return (a/(a+m) * b) + (m/(m+a) * c)
`````

`````python
depensesMax['CBA'] = depensesMax.apply(weighted_rating, axis=1)
`````

`````python
# Sort by calculating bayesian average

depensesMax = depensesMax.sort_values('CBA', ascending=False)

depensesMax.head()
`````

Top rating by Secteur


`````python
# Merge the depensesMax dataset with nbrA on id_client
top_secteur = pd.merge(depensesMax, nbrA, how='left', left_on='id_client', right_on='id_client')

top_secteur.head(20)

`````

`````python
def top_by_secteur(Secteur, n):
    filtered_rows = top_secteur[top_secteur['Secteur'] == Secteur]
    sorted_rows = filtered_rows.sort_values(by='CBA', ascending=False)
    top_n_clients = sorted_rows.head(n)
    return top_n_clients

`````

**Test the similarity table results**

**Sim table with client similarity**

`````python
# Calculate similarities
pivotA = nbrA.pivot_table(index='id_client', columns='nom_carte_fidelite', values='Nombre_A_ou_MA')
pivotA = pivotA.fillna(0)

pivotA.head()
`````

`````python
model_knn = NearestNeighbors(metric='cosine')
model_knn.fit(pivotA)
`````

`````python
distances, indices = model_knn.kneighbors(pivotA, n_neighbors=5)
clt_sim = pd.DataFrame(indices, columns=['client0', 'client1', 'client2', 'client3', 'client4'])
`````

`````python
# Merge clt_sim with the aggregated data
client_ = clt_sim.copy()
for i in range(0, 5):
    similar_client_indices = clt_sim[f'client{i}']
    similar_client_names =pivotA.index[similar_client_indices]
    client_[f'client{i}'] = similar_client_names
client_.head(10)
`````

`````python
client_id_to_name = dict(zip(nbrA['id_client'], nbrA['nom_prenom']))

# Replace client IDs in the similarity matrix with names
for i in range(5):  # Replace '5' with the actual number of clients in each row
    client_.iloc[:, i] = client_.iloc[:, i].map(client_id_to_name)
`````
